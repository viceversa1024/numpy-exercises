#### 42. Consider two random arrays A and B, check if they are equal (★★☆)

````python
A = np.random.rand(25).reshape((5,5))
B = np.random.rand(25).reshape((5,5))


def check_equal(A, B):
    if A.shape == B.shape:
        if np.all(A == B):
            return True
        else:
            return False
    else:
        return print("not the same shape")
    
check_equal(A, B)
````

#### 44. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

````python
A = np.linspace(1,20,20).reshape((10,2))
def cart_to_polar(a):
    for i in a:
        x = i[0]
        y = i[1]
        r = np.sqrt((x*x)+(y*y))
        theta = np.arctan(y/x)
        print(r, theta)

cart_to_polar(A)
````
